```python
# Map labels to proper words
mappings = {'fh': 'forehand',
           'bh': 'backhand',
           'CC': 'cross-court',
           'DL': 'down the line',
           'DM': 'down the middle',
           'IO': 'inside out',
           'II': 'inside in',
           ' W ': ' wide ',
           ' B ': ' body '
           }

    
```

```python
all_json = []

def write_to_dict(cls, name):
    to_write = {}
    to_write['id'] = len(all_json)
    to_write['video'] = cls + '/' + name + ".mp4"
#     if not os.path.isfile('D:/fyp/datasets/' + to_write['video']):
#         return
    to_write['conversations'] = [{
        "from": "human", 
        "value": "<video>\nWhat tennis shot is the person in the video hitting?"
    },
    {
        "from": "gpt",
        "value": "The person is hitting a " + ' '.join(cls.split('_'))
    }]
    
    for i in mappings:
        to_write['conversations'][1]["value"] = to_write['conversations'][1]["value"].replace(i, mappings[i])

    all_json.append(to_write)
```

```python
for index, row in train.iterrows():
    events = row['events']
    name = row['video']
    for event in events:
        label = event['label']
        segmented_vid = name.split('_')
        frame = int(event['frame'])
        combined = name + "_" + str(frame)
        write_to_dict(label, combined) 

print(all_json[0])
```
